```python
df1 = pd.read_csv('https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv', index_col=0, parse_dates=True)
df1.info()
```

```python
df1['WeeklyDay'] = df1.index.day_name()
df1.head()
```

```python
df1.loc['2017'][['Consumption', 'Wind', 'Solar']].plot(subplots=True, figsize=(11, 4))
plt.show()
```

```python
df1['Consumption'].plot(linewidth=0.5)
plt.show()
```

```python
df1.loc['2017'][['Consumption', 'Wind', 'Solar']].resample('Q').sum().plot()
plt.show()
```
