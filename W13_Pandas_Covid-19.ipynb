```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.drop(['day', 'month', 'year', 'countryterritoryCode', 'popData2019', 
         'Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'], axis=1, inplace=True)
df
```

```python
df1 = df[['cases', 'continentExp']].groupby('continentExp').sum()
df1.sort_values(by=['cases'], inplace=True, ascending=False)
df1
```

```python
df1.plot.pie(y='cases',figsize=(10, 20), autopct='%1.1f%%')
```

```python
dfx = df.groupby('continentExp').agg(['max', 'min', 'mean'])
dfx
```
