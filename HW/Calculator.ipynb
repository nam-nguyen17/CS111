```python
from tkinter import *
from PIL import ImageTk, Image
```

```python
expression = ""

# Function to update expression in the Entrybox
def digit(num):
    global expression
    expression = expression + str(num)
    equation.set(expression)    # update the expression

# Function to evaluate 
def calculate():
    try:
        global expression
        total = str(eval(expression))
        equation.set(total)
        expression = total         

    except:     # error when 1/0, ... or words or type in the keyboard
        equation.set(" ERROR ") 
        expression = ""

# Function to clear 
def clear():
    global expression
    expression = ""
    equation.set("")
    

if __name__ == "__main__":
    root = Tk()
    root.title("Calculator")
    root.configure(background="light green")
    root.resizable(0, 0)    # Restricting root window

    # hold string values
    equation = StringVar()

    # create Entrybox to show the expression
    expression_field = Entry(root, textvariable=equation, bd=10, bg="#2f6212", justify="right", font=("Arial", 20))
    expression_field.grid(row=0, columnspan=5, ipadx=60)

    # create a Buttons and place
    # Button 1-2-3
    Button(root, text="1",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(1)).grid(row = 3, column = 0)
    Button(root, text="2",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(2)).grid(row = 3, column = 1)
    Button(root, text="3",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(3)).grid(row = 3, column = 2)

    # Button 4-5-6
    Button(root, text="4",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(4)).grid(row = 2, column = 0)
    Button(root, text="5",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(5)).grid(row = 2, column = 1)
    Button(root, text="6",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(6)).grid(row = 2, column = 2)

    # Button 7-8-9-0
    #Button(root, text="7", padx=35, pady=20, command= lambda: digit(7)).grid(row = 1, column = 0)
    Button(root, text="7",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(7)).grid(row = 1, column = 0)

    Button(root, text="8",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(8)).grid(row = 1, column = 1)
    Button(root, text="9",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(9)).grid(row = 1, column = 2)
    Button(root, text="0",font=("Arial", 12, "bold"), bd= 12, height=2, width=6, command= lambda: digit(0)).grid(row=4, column=0)

    # Button "." and =
    Button(root, text=".",font=("Arial", 12, "bold"), bd= 12,bg='#9b9d8f', height=2, width=6, command= lambda: digit(".")).grid(row = 3, column = 3)
    Button(root, text="=",font=("Arial", 12, "bold"), bd= 12,bg='#9b9d8f', height=2, width=6, command=calculate).grid(row = 3, column = 4)

    # Button +-*/
    Button(root, text="+",font=("Arial", 12, "bold"), bd= 12, bg='#9b9d8f',height=2, width=6, command= lambda: digit("+")).grid(row = 2, column = 3)
    Button(root, text="-",font=("Arial", 12, "bold"), bd= 12, bg='#9b9d8f',height=2, width=6, command= lambda: digit("-")).grid(row = 2, column = 4)
    Button(root, text="*",font=("Arial", 12, "bold"), bd= 12, bg='#9b9d8f',height=2, width=6, command= lambda: digit("*")).grid(row = 1, column = 3)
    Button(root, text="/",font=("Arial", 12, "bold"), bd= 12, bg='#9b9d8f',height=2, width=6, command= lambda: digit("/")).grid(row = 1, column = 4)

    # Button Clear
    Button(root, text="Clear",font=("Arial", 12, "bold"), bd= 12, bg='orange', height=2, width=33, command=clear).grid(row=4, column=1,columnspan=5)

    # start the ROOT
    root.mainloop()
```
