```python
def show_folder(folder, space=''):
#     print(space+folder)
    contents = os.listdir(folder)
    for fn in contents:
#         print(space + '+----' + fn)
        fullname = os.path.join(folder, fn)  # folder + '/' + fn
        if os.path.isfile(fullname):
            print(space + '+----' + fn)
        elif os.path.isdir(fullname):
            print(space + '\----' + fn)
            show_folder(fullname, space + '|    ')
```

```python
show_folder('/Users/Storage/TTU/Course/')
```

```python
def show_folder2(folder, level=0):
    if level == 0:
        print('ROOT:' + folder)
    space = '|    ' * (level + 1)
    contents = os.listdir(folder)
    for fn in contents:
        fullname = os.path.join(folder, fn) 
        if os.path.isfile(fullname):
#             print("{}{}{}".format(space,'+----',fn))
            print(space + '+----' + fn)
        elif os.path.isdir(fullname):
            print("{}{}{}".format(space,'\----',fn))
            show_folder2(fullname, level + 1)
```

```python
from collections import defaultdict
n = int(input())
m = int(input())
d = defaultdict(list)
A = []
for i in range(0, n):
    #i = input()
    d[input()].append(i)
print(d)

for j in range(0, m):
    A = A + [input()] 
print(A)

for i in A:
    if i in d:
        print(' '.join(map(str, d[i])))
    else:
        print(-1)
```
